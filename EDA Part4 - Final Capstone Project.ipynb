```python
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import daily prices dataset using Pandas
stock_daily_prices_df = pd.read_csv('data/stock_daily_prices.csv')
stock_daily_prices_df.head(3)
```

```python
# Use matplotlib on the Pandas DataFrame to plot the data
stock_daily_prices_df.plot(x = 'Date', linewidth = 3, figsize = (14, 6));
plt.ylabel('Price [$]')
plt.title('All 3 stocks daily prices')
plt.legend(loc = "upper left")
plt.grid()
```

```python
# Using Matplotlib, plot 3 stocks daily prices on multiple subplots
stock_daily_prices_df.plot(x = 'Date', linewidth = 3, subplots = True, figsize = (14, 6));
plt.ylabel('Price [$]')
plt.title('All 3 stocks daily prices')
plt.legend(loc = "upper left")
plt.grid()
```

```python
# Using Matplotlib, plot the 3 plots on subplots next to each other (all figures in one row).
plt.figure(figsize = (20, 10))

plt.subplot(1, 3, 1)
plt.plot(stock_daily_prices_df['NFLX'], 'r');
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(stock_daily_prices_df['FB'], 'b');
plt.grid()


plt.subplot(1, 3, 3)
plt.plot(stock_daily_prices_df['TWTR'], 'g');
plt.grid()
```

```python
# Import daily returns dataset using Pandas
stocks_daily_returns_df = pd.read_csv('data/stocks_daily_returns.csv')
stocks_daily_returns_df.head(3)
```

```python
# Using Matplotlib, plot the scatterplot between Facebook and Twitter daily returns.
stocks_daily_returns_df.plot.scatter('FB', 'TWTR', grid = True, figsize = (12, 7));
```

```python
# Using Seaborn, plot similar scatterplot between Facebook and Twitter daily returns.
plt.figure(figsize = (15, 8))
sns.scatterplot(x = 'FB', y = 'TWTR', data = stocks_daily_returns_df);
plt.grid()
```

```python
# Assume that you now expanded your portfolio to include additional stocks such as Amazon (AMZN) and Google (GOOG).
# You decided to become bullish on Twitter and you allocated 60% of your assets in it.
# You also decided to equally divide the rest of your assets in other stocks (AMZN, FB, GOOG, NFLX).
# Using Matplotlib, plot a pie chart that shows these allocations. Use 'explodeâ€™ attribute to increase the separation between TWTR and the rest of the portfolio.

allocation_dict =  {'allocation %': [60, 10, 10, 10, 10]}
stock_df = pd.DataFrame(data = allocation_dict, index = ['TWTR', 'FB', 'NFLX', 'AMZN', 'GOOG'])

explode = (0.2, 0, 0, 0, 0)
stock_df.plot.pie(y = 'allocation %', figsize = (8, 8), explode=explode)
plt.title('PORTFOLIO ALLOCATIONS PIE CHART')
```

```python
# Using Matplotlib, plot the histogram for FB returns using 40 bins with red color. Display the mean and Standard deviation on top of the figure.

mu = stocks_daily_returns_df['FB'].mean() # mean of distribution
sigma = stocks_daily_returns_df['FB'].std() # standard deviation of distribution
 
num_bins = 40

# Plot the histogram of the returns
plt.figure(figsize = (12, 6))
plt.hist(stocks_daily_returns_df['FB'], num_bins, facecolor = 'red');
plt.grid()
plt.ylabel('Returns ($)')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));

 
```

```python
# Using Seaborn, plot a heatmap that shows the correlations between stocks daily returns.
plt.figure(figsize = (12, 6)) 
sns.heatmap(stocks_daily_returns_df.corr(), annot = True) 
```
